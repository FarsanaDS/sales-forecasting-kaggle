### Feature Engineering

```python
sales_train['year'] = sales_train['date'].dt.year
sales_train['month'] = sales_train['date'].dt.month
sales_train['day'] = sales_train['date'].dt.day
sales_train['weekday'] = sales_train['date'].dt.weekday  # 0 = Monday, 6 = Sunday
sales_train['week_of_year'] = sales_train['date'].dt.isocalendar().week
sales_train['quarter'] = sales_train['date'].dt.quarter
sales_train['is_weekend'] = (sales_train['weekday'] >= 5).astype(int)  # 1 if Sat/Sun, else 0
```

```python
sales_train['sales_lag_1'] = sales_train['sales'].shift(1)
sales_train['sales_lag_7'] = sales_train['sales'].shift(7)  # Sales 1 week ago
sales_train['sales_lag_30'] = sales_train['sales'].shift(30)  # Sales 1 month ago
```

```python
sales_train['sales_ma_7'] = sales_train['sales'].rolling(window=7).mean()
sales_train['sales_ma_30'] = sales_train['sales'].rolling(window=30).mean()
```

```python
sales_train['sales_ema_7'] = sales_train['sales'].ewm(span=7, adjust=False).mean()
sales_train['sales_ema_30'] = sales_train['sales'].ewm(span=30, adjust=False).mean()
```

```python
sales_train['price_change'] = sales_train['sell_price_main'].pct_change()
sales_train['price_elasticity'] = sales_train['sales'] / sales_train['sell_price_main']
```

```python
sales_train['is_holiday'] = sales_train['holiday']
sales_train['is_school_holiday'] = sales_train['school_holidays']
sales_train['is_winter_holiday'] = sales_train['winter_school_holidays']
```

```python
discount_cols = ['type_0_discount', 'type_1_discount', 'type_2_discount',
                 'type_3_discount', 'type_4_discount', 'type_5_discount', 'type_6_discount']

sales_train['total_discount'] = sales_train[discount_cols].sum(axis=1)
sales_train['discount_rate'] = sales_train['total_discount'] / (sales_train['sell_price_main']+1e-9)
```

```python
sales_train['is_out_of_stock'] = (sales_train['availability']==0).astype(int)
```

```python
warehouse_sales = sales_train.groupby('warehouse')['sales'].agg(['mean', 'median', 'std', 'max', 'min']).reset_index()
sales_train = sales_train.merge(warehouse_sales, on='warehouse',how="left")
```

```python
sales_train['sales_per_order'] = sales_train['sales'] / (sales_train['total_orders']+1e-9)
sales_train['discount_per_order'] = sales_train['total_discount'] / (sales_train['total_orders']+1e-9)
sales_train['sales_per_availability'] = sales_train['sales'] / (sales_train['availability']+1e-9)
```

```python
for col in sales_train.select_dtypes(include=['float64']).columns:
    sales_train[col] = sales_train[col].astype('float32')  # Reduces memory usage

for col in sales_train.select_dtypes(include=['int64']).columns:
    sales_train[col] = sales_train[col].astype('int32')  # Reduces memory usage
```

```python
sales_train['date'] = sales_train['date'].astype('datetime64[ns]').astype(int)

sales_train = sales_train.apply(pd.to_numeric, errors='coerce')
```

```python
sales_train.fillna(sales_train.mean(), inplace=True)
```

```python
sales_test=sales_test.merge(inventory,on="unique_id",how='left')
sales_test=sales_test.merge(calendar,on=["date","warehouse"],how="left")
```

```python
sales_test=sales_test.drop(columns=["warehouse_x","warehouse_y","product_unique_id_y","name_y","L1_category_name_en_y",
                                      "L2_category_name_en_y","L3_category_name_en_y","L4_category_name_en_y"])
```

```python
#rename remaining column for clarity
sales_test=sales_test.rename(columns={"product_unique_id_x":"product_unique_id","name_x":"product_name",
                                       "L1_category_name_en_x":"L1_category","L2_category_name_en_x":"L2_category",
                                       "L3_category_name_en_x":"L3_category","L4_category_name_en_x":"L4_category"})
```

```python
encoder=LabelEncoder()
sales_test['warehouse']=encoder.fit_transform(sales_test['warehouse'])
```

```python
sales_test=pd.get_dummies(sales_test,columns=["L1_category","L2_category","L3_category","L4_category"], drop_first=True)
```

```python
sales_test['product_name']=LabelEncoder().fit_transform(sales_test['product_name'])
```

```python
# Extract date-based features
sales_test['year'] = sales_test['date'].dt.year
sales_test['month'] = sales_test['date'].dt.month
sales_test['day'] = sales_test['date'].dt.day
sales_test['weekday'] = sales_test['date'].dt.weekday
sales_test['week_of_year'] = sales_test['date'].dt.isocalendar().week
sales_test['quarter'] = sales_test['date'].dt.quarter
sales_test['is_weekend'] = (sales_test['weekday'] >= 5).astype(int)

# Sales-related features (fill with averages from sales_train)
sales_test['sales_lag_1'] = sales_train['sales_lag_1'].mean()
sales_test['sales_lag_7'] = sales_train['sales_lag_7'].mean()
sales_test['sales_lag_30'] = sales_train['sales_lag_30'].mean()
sales_test['sales_ma_7'] = sales_train['sales_ma_7'].mean()
sales_test['sales_ma_30'] = sales_train['sales_ma_30'].mean()
sales_test['sales_ema_7'] = sales_train['sales_ema_7'].mean()
sales_test['sales_ema_30'] = sales_train['sales_ema_30'].mean()

# Price-related features
sales_test['price_change'] = sales_test['sell_price_main'].pct_change().fillna(0)
sales_test['price_elasticity'] = 1  # Placeholder, as sales is missing

# Holiday features
sales_test['is_holiday'] = sales_test['holiday']
sales_test['is_school_holiday'] = sales_test['school_holidays']
sales_test['is_winter_holiday'] = sales_test['winter_school_holidays']

# Discount-related features
discount_cols = ['type_0_discount', 'type_1_discount', 'type_2_discount',
                 'type_3_discount', 'type_4_discount', 'type_5_discount', 'type_6_discount']

sales_test['total_discount'] = sales_test[discount_cols].sum(axis=1)
sales_test['discount_rate'] = sales_test['total_discount'] / (sales_test['sell_price_main'] + 1e-9)

# Out of stock feature
sales_test['is_out_of_stock'] = 0  # We assume it's available (since availability is missing)

# Aggregate sales-related features from train and merge with test
warehouse_sales = sales_train.groupby('warehouse')['sales'].agg(['mean', 'median', 'std', 'max', 'min']).reset_index()
sales_test = sales_test.merge(warehouse_sales, on='warehouse', how='left')

# Fill missing values for other sales-based features
sales_test['sales_per_order'] = sales_train['sales_per_order'].mean()
sales_test['discount_per_order'] = sales_train['discount_per_order'].mean()
sales_test['sales_per_availability'] = sales_train['sales_per_availability'].mean()
```

```python
# Compute the average availability per product from sales_train
avg_availability = sales_train.groupby('unique_id')['availability'].mean().reset_index()
avg_availability.rename(columns={'availability': 'estimated_availability'}, inplace=True)

# Merge the estimated availability with sales_test
sales_test = sales_test.merge(avg_availability, on='unique_id', how='left')

# Fill missing values (some products in test may not exist in train)
sales_test['estimated_availability'].fillna(1, inplace=True)  # Assume fully available if unknown

# Rename to match train dataset column name
sales_test.rename(columns={'estimated_availability': 'availability'},inplace=True)
```

```python
for col in sales_test.select_dtypes(include=['float64']).columns:
    sales_test[col] = sales_test[col].astype('float32')  # Reduces memory usage

for col in sales_test.select_dtypes(include=['int64']).columns:
    sales_test[col] = sales_test[col].astype('int32')  # Reduces memory usage
```

```python
sales_test['date'] = sales_test['date'].astype('datetime64[ns]').astype(int)

sales_test = sales_test.apply(pd.to_numeric, errors='coerce')
```

```python
sales_test.fillna(sales_test.mean(), inplace=True)
```

```python
# Ensure test set has all the same columns as train set
missing_cols = set(sales_train.columns) - set(sales_test.columns)

# Add missing columns to test with default value 0
for col in missing_cols:
    sales_test[col]=0
```

```python
common_cols = list(set(sales_train.columns) & set(sales_test.columns))
sales_train = sales_train[common_cols]
sales_test = sales_test[common_cols]
```

```python
# Get feature names of the training data
train_features = set(sales_train.columns)

# Get feature names of the test data
test_features = set(sales_test.columns)

# Identify mismatches
missing_in_test = train_features - test_features
extra_in_test = test_features - train_features

print("Features in training but missing in test:", missing_in_test)
print("Features in test but missing in training:", extra_in_test)
```

```python
x=sales_train.drop('sales',axis=1)
y=sales_train['sales']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25, random_state=42)
```

```python
import gc
import sys
gc.collect()
```

```python
import xgboost as xgb
from sklearn.model_selection import RandomizedSearchCV
import numpy as np

# Define the model
xgb_model = xgb.XGBRegressor(objective='reg:squarederror', random_state=42)

# Define hyperparameter distributions
param_dist = {
    'n_estimators': np.arange(50, 200, 50),
    'max_depth': [3, 6],
    'learning_rate': np.linspace(0.01, 0.2, 5),
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.8, 1.0],
    'gamma': [0, 0.1],
    'reg_alpha': [0, 0.5],
    'reg_lambda': [0.5, 1.0]
}

# Perform Randomized Search
random_search = RandomizedSearchCV(estimator=xgb_model, param_distributions=param_dist, 
                                   n_iter=5, cv=2, n_jobs=1, scoring='neg_mean_absolute_error')

# Fit on training data
random_search.fit(x_train, y_train)

# Best parameters
print("Best Parameters:", random_search.best_params_)

# Use best model
best_xgb = random_search.best_estimator_
```

```python
train_pred = best_xgb.predict(x_train)
```

```python
r2_score(train_pred,y_train)
```

```python
test_pred = best_xgb.predict(x_test)
```

```python
r2_score(test_pred,y_test)
```

```python
import numpy as np

def WMAE(y_test,test_pred,weights):
    return np.sum(weights*np.abs(y_test-test_pred))/np.sum(weights)
    
score=WMAE(y_test,test_pred,weights=x_test['availability'].values)
print(score)
```
